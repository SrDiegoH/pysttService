```python
#!flask/bin/python
import re
import base64
import speech_recognition as sr
from flask import Flask, request, json
```

```python
app = Flask(__name__)
```

```python
@app.route('/')
def root():
    return 'Working'
```

```python
@app.route('/stt', methods=['POST'])
def stt():
    if(request.is_json):
        content = request.get_json()
        return to_text(content['data'])
    return 'Error'
```

```python
def process_audio(audio_file):
    r = sr.Recognizer()
    
    text = ''
    with sr.AudioFile(audio_file) as source:
        audio = r.record(source)
        try:
            text = r.recognize_sphinx(audio)
        except Exception as e:
            text = f'Error: {e}'
    return text
```

```python
def to_text(string_as_base64_data):
    return process_audio(convert_base64_to_file(string_as_base64_data))
```

```python
#file_name = 'F:\\Sistema\\Downloads\\bigdatamusic_datas\\audios\\file.wav'
file_name = 'file.wav'
```

```python
def convert_base64_to_file(string_as_base64_data):
    string_as_base64_data = string_as_base64_data + '=' * (-len(string_as_base64_data) % 4)
    
    audio_file = open(file_name, "wb")
    audio_file.write(base64.b64decode(string_as_base64_data))
    audio_file.close()
    
    return file_name
```
